```python
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score
```

```python
metode_list = ["cosine", 'dice', 'jaccard']
praproses_list = ["True", "False"]
target_names = ["1", "0", "9"]

precision_true = list()
precision_false = list()
recal_true = list()
recal_false = list()
f_score_true = list()
f_score_false = list()
akurasi_true = list()
akurasi_false = list()

for praproses in praproses_list:
    for metode in metode_list:
        data = pd.read_excel("hasil skor/hasil_"+metode+"_praproses_"+praproses+".xlsx")
        prediksi = data['prediksi_'].tolist()
        asli = data['label asli'].tolist()
        y_true = [str(x) for x in asli]
        y_pred = [str(x) for x in prediksi]
        print("Metode:",metode,"||  ", "Praproses:",praproses)
        print(classification_report(y_true, y_pred, target_names=target_names))
        print("="*100)
        print("")
        
        if praproses == "True":
            hasil = list(precision_recall_fscore_support(y_true, y_pred, average='weighted'))
            precision_true.append(hasil[0])
            recal_true.append(hasil[1])
            f_score_true.append(hasil[2])
            akurasi_true.append(accuracy_score(y_true, y_pred))
        else:
            hasil = list(precision_recall_fscore_support(y_true, y_pred, average='weighted'))
            precision_false.append(hasil[0])
            recal_false.append(hasil[1])
            f_score_false.append(hasil[2])
            akurasi_false.append(accuracy_score(y_true, y_pred))
    print("*"*100)
```

# Kesimpulan

```python
dict_ = {
    "metode":metode_list,
    'precision_false' :precision_false,
    'recal_false' :recal_false,
    'f_score_false' :f_score_false,
    'akurasi_false':akurasi_false,
    
    "|":['|'for i in range(len(akurasi_true))],
    
   'precision_true' :precision_true,
    'recal_true' :recal_true,
    'f_score_true' :f_score_true,
    'akurasi_true' :akurasi_true,
}
perbandingan1 = pd.DataFrame.from_dict(dict_)
perbandingan1
```
