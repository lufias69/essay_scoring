```python
from ES import scoring as sc
import pandas as pd
```

```python
#Jawaban
dir_ = 'data'
file_name = 'jawaban'
data = pd.read_excel(dir_+"/"+file_name+".xlsx")
jawaban = data[:100]["jawaban"].tolist()
label_asli = data[:100]["score"].tolist()
print(len(jawaban))
data.head()
```

```python
kunci_jawaban = [
                 "telepon 02173811111 website buanwisata.com",
                 "02173811111 buanawiata.com",
                ]
skor = [1,1]
```

# Cosine 

```python
metode = "cosine"
praproses = True

similarity = list()
skor_prediksi = list()
jawaban_clean  = list()
bobot_list = list()
for jawaban_ in jawaban:
    output = sc.scoring(kunci_jawaban,skor,jawaban=jawaban_, metode = metode, praproses=praproses)
    similarity.append(output[0])
    skor_prediksi.append(output[1])
    bobot_list.append(output[-2])
    jawaban_clean.append(output[-1])

dict1 = {
    'similarity':similarity,
    'prediksi_':skor_prediksi,
    'label asli':label_asli,
    'jawaban':jawaban,
    'jawaban_clean':jawaban_clean
}
dict_bobot = {
    'similarity':similarity,
    'prediksi_':skor_prediksi,
    'label asli':label_asli,
    'jawaban':jawaban,
    'jawaban_clean':jawaban_clean,
    'bobot':bobot_list
}
df = pd.DataFrame.from_dict(dict1)
df.to_excel("hasil skor/hasil_"+metode+"_praproses_"+str(praproses)+".xlsx")
df
```
