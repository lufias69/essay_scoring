```python
from ES import scoring as sc
import pandas as pd
```

```python
#Jawaban
dir_ = 'data'
file_name = 'jawaban'
data = pd.read_excel(dir_+"/"+file_name+".xlsx", sheet_name= "Sheet3")
jawaban = data["jawaban"].tolist()
label_asli = data["skor"].tolist()
# print(len(jawaban))
data.head(10)
```

```python
kunci_jawaban = [
                 "telepon 02173811111 website buanwisata.com",
                 "02173811111 buanawisata.com",
                 "menelpon mengunjungi website",
                "mengunjungi situs web buana wisata menelpon",
                "brosur tercantum nomor telopon brosur tercantum website dapat dikunjungi",
                "dengan menggunjungi website dengan nomer telepon",
                "dengan menelpon mengunjungi website yang ada"
                ]
skor = [1,1,1,1,1,1,1]

pd.DataFrame.from_dict({
    'skor':skor,
    "kunci_jawaban":kunci_jawaban,
})
```

# Cosine 

```python
metode = "cosine"
praproses = True

similarity = list()
skor_prediksi = list()
jawaban_clean  = list()
bobot_list = list()
for ix, jawaban_ in enumerate(jawaban):
    output = sc.scoring(kunci_jawaban,skor,jawaban=jawaban_, metode = metode, praproses=praproses)
    similarity.append(output[0])
    skor_prediksi.append(output[1])
    bobot_list.append(output[-2])
    jawaban_clean.append(output[-1])
    if (ix+1)%100==0:
        print(ix)
    else:
        print(".", end="")

dict1 = {
    'similarity':similarity,
    'prediksi_':skor_prediksi,
    'label asli':label_asli,
    'jawaban':jawaban,
    'jawaban_clean':jawaban_clean
}
dict_bobot = {
    'similarity':similarity,
    'prediksi_':skor_prediksi,
    'label asli':label_asli,
    'jawaban':jawaban,
    'jawaban_clean':jawaban_clean,
    'bobot':bobot_list
}
df = pd.DataFrame.from_dict(dict1)
df.to_excel("hasil skor/hasil_"+metode+"_praproses_"+str(praproses)+".xlsx")
df
```

```python
metode = "dice"
praproses = True

similarity = list()
skor_prediksi = list()
jawaban_clean  = list()
bobot_list = list()
for ix, jawaban_ in enumerate(jawaban):
    output = sc.scoring(kunci_jawaban,skor,jawaban=jawaban_, metode = metode, praproses=praproses)
    similarity.append(round(output[0],2))
    skor_prediksi.append(output[1])
    bobot_list.append(output[-2])
    jawaban_clean.append(output[-1])
    if (ix+1)%100==0:
        print(ix+1)
    else:
        print(".", end="")

dict1 = {
    'similarity':similarity,
    'prediksi_':skor_prediksi,
    'label asli':label_asli,
    'jawaban':jawaban,
    'jawaban_clean':jawaban_clean
}
dict_bobot = {
    'similarity':similarity,
    'prediksi_':skor_prediksi,
    'label asli':label_asli,
    'jawaban':jawaban,
    'jawaban_clean':jawaban_clean,
    'bobot':bobot_list
}
df = pd.DataFrame.from_dict(dict1)
df.to_excel("hasil skor/hasil_"+metode+"_praproses_"+str(praproses)+".xlsx")
df
```
