```python
from ES import scoring as sc
import pandas as pd
```

```python
#Jawaban
dir_ = 'data'
file_name = 'jawaban'
data = pd.read_excel(dir_+"/"+file_name+".xlsx")
jawaban = data[:100]["jawaban"].tolist()
print(len(jawaban))
data.head()
```

```python
jawaban[89]
```

# Kunci Jawaban

```python
kunci_jawaban = ["telepon 02173811111 website buanwisata.com",
                 "02173811111 buanawiata.com",
                ]
```

# Implementasi Cosine Similarity

```python
prediksi_similarity = list()
prediksi_ = list()
jawaban_clean  = list()
for jawaban_ in jawaban:
    output = sc.predict(kunci_jawaban, jawaban_, metode='cosine', praproses=False)
    prediksi_similarity.append(output[0])
    prediksi_.append(output[1])
    jawaban_clean.append(output[2])

dictt = {
    'similarity':prediksi_similarity,
    'prediksi_':prediksi_,
    'label asli':data[:100]["Column4"].tolist(),
    'jawaban':jawaban,
    'jawaban_clean':jawaban_clean
}
df = pd.DataFrame.from_dict(dictt)
df
```
